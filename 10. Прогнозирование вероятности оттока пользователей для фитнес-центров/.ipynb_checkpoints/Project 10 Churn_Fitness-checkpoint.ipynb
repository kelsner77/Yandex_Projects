# Прогнозирование вероятности оттока пользователей для фитнес-центров

**Описание проекта**

Проект направлен на разработку стратегии взаимодействия с клиентами сети фитнес-центров «Культурист-датасаентист» на основе аналитических данных с целью их удержания. Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал.

В рамках проекта будут реализованы следующие задачи:

*    прогноз вероятности оттока (на уровне следующего месяца) для каждого клиента;
*    формирование типичных портретов клиентов: выделение несколько наиболее ярких групп и их основных свойств;
*    анализ основных признаков, наиболее сильно влияющих на отток;
*    формулирование основных выводов и разработка рекомендаций по повышению качества работы с клиентами:
        1. выделение целевых групп клиентов;
        2. предложение мер по снижению оттока;
        3. определение других особенностей взаимодействия с клиентами.

Шаг 1. Исследовательский анализ данных (EDA)

*    Наличие в датасете отсутствующих признаков, изучение средних значений и стандартных отклонений (метод describe()).
*    Средние значения признаков в двух группах — тех, кто ушел в отток и тех, кто остался (методом groupby()).
*    Создание столбчатых гистограмм и распределения признаков для тех, кто ушёл (отток) и тех, кто остался (не попали в отток).
*    Создание матрицы корреляций.


Шаг 2. Создание модели прогнозирования оттока клиентов

*    Разделение данных на обучающую и валидационную выборку функцией train_test_split().
*    Обучение модели на train-выборке двумя способами: логистической регрессией, случайным лесом.
*    Оценка метрики accuracy, precision и recall для обеих моделей на валидационной выборке. Сравнение по ним модели. Какая модель показала себя лучше на основании метрик?

Шаг 3. Кластеризация клиентов

*    Стандартизация данных.
*    Создание матрицы расстояний функцией linkage() на стандартизованной матрице признаков и построение дендрограммы. Количество кластеров, которые можно выделить.
*    Обучение модели кластеризации на основании алгоритма K-Means и прогнозирование кластеров клиентов.
*    Изучение средних значений признаков для кластеров. 
*    Построение распределения признаков для кластеров. 
*    Расчет доли оттока (методом groupby()) для каждого полученного кластера. Наличие отличий по доле оттока. Какие кластеры склонны к оттоку, а какие — надёжны?

Шаг 4. Выводы и базовые рекомендации по работе с клиентами

* Функция для изучения общей информации о таблице

```python
def description(x):
    print('\033[1m' + 'Получение общей информации о данных в таблице:' + '\033[0m')
    display(x.info())
    print('\033[1m' + 'Количество строк и столбцов:' + '\033[0m')
    display(x.shape)
    print('\033[1m' + 'Подсчёт пропусков в таблице:' + '\033[0m')
    display(pd.concat([x.isna().sum(), x.isna().mean()], axis=1)\
    .rename(columns={0:'кол-во', 1:'доля'})\
    .style.format({'кол-во':'{:.0f}', 'доля':'{:.2%}'}))
    print('\033[1m' + 'Наличие дубликатов в таблице:' + '\033[0m')
    display(x.duplicated().sum())
    print('\033[1m' + 'Доля дубликатов в таблице:' + '\033[0m')
    display("{:.3%}".format(x.duplicated().sum() / len(x), 2))
    print('\033[1m' + 'Уникальные значения в таблице:' + '\033[0m')
    display(x.nunique())
    print('\033[1m' + 'Статистические показатели таблицы:' + '\033[0m')
    display(x.describe())
```

*        'partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
*        'promo_friends' — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
*        'phone' — наличие контактного телефона;
*        'age' — возраст;
*        'lifetime' — время с момента первого обращения в фитнес-центр (в месяцах).
    
Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
*        'contract_period' — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
*        'month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах);
*        'group_visits' — факт посещения групповых занятий;
*        'avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента;
*        'avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц;
*        'avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
*        'churn' — факт оттока в текущем месяце.

В столбцах gender, near_location, partner, promo_friends, phone, group_visits и churn только по два уникальных значения - 0 и 1. В столбце contract_period - 3 уникальных значения (6, 12 и 1). Столбец month_to_end_contract содержит 12 уникальных значений. Данные соответствуют описанию таблицы.

В таблице отсутствуют пропущенные значения и дубликаты. Типы данных соответствуют значениям. Заглавные буквы названий колонок заменены на строчные.

Средний портрет клиента - это 29-летний человек, который выбирает фитнес-центр поближе к дому или работе, посещает его почти 2 раза в неделю и приносит выручку от дополнительных услуг в размере 146,94 у.е.

Чаще всего клиенты уходят, если не являются сотрудниками компании-партнера, немного моложе, приходят не по рекомендациям знакомых, берут абонемент на более короткий срок и реже посещают групповые и обычные занятия.

## Создание модели прогнозирования оттока клиентов

Модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце.

###    Разделение данных на обучающую и валидационную выборку функцией train_test_split()

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop(columns=['churn'])
y = df['churn']
```

```python
# указываем параметр random_state при разделении выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Параметру random_state передано значение для того, чтобы результат работы алгоритма можно было воспроизвести при перезапуске. 

###    Обучение модели на train-выборке двумя способами: логистической регрессией, случайным лесом.

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
```

```python
# обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_train_st = scaler.fit_transform(X_train)
```

```python
# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

Применена стандартизация данных, чтобы свести признаки к одному масштабу.

###    Оценка метрики accuracy, precision и recall для обеих моделей на валидационной выборке. Сравнение по ним модели.

```python
# список моделей
models = [
    LogisticRegression(solver='liblinear', random_state=0),
    RandomForestClassifier(random_state=0)
]
```

Для отключения предупреждения библиотеки sklearn о том, что в будущем алгоритм решения по умолчанию, так называемый солвер, будет изменён, добавлен аргумент solver='liblinear'.

```python
# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(
        'Accuracy: {:.2f} Precision: {:.2f} Recall: {:.2f}'.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred),
            recall_score(y_test, y_pred)
        )
    )
```

```python
# цикл, который выводит метрики по списку моделей
for i in models:
    print(i)
    make_prediction(i, X_train_st, y_train, X_test_st, y_test)
```

Метрика accuracy означает долю верно угаданных ответов из всех прогнозов. Чем ближе значение accuracy к 100%, тем лучше. Работает только при условии баланса классов — когда объектов каждого класса примерно поровну, 50% : 50%. 

Precision показывает долю правильных ответов только среди целевого класса. 

Recall говорит о том, сколько реальных объектов целевого класса мы обнаружили с помощью модели.

**Вывод**

Созданы две модели прогнозирования оттока клиентов: LogisticRegression и RandomForestClassifier. По трем метрикам сравнены модели для определения лучшей: accuracy, precision и recall.

В данном исследовании модель "логистическая регрессия" показала себя немного лучше "Случайного леса" на основании метрик "Precision" и "Recall". Значение метрики "Accuracy" почти одинаково у обеих моделей.

## Кластеризация клиентов

###   Стандартизация данных

```python
# стандартизируем данные
sc = StandardScaler()
x_sc = sc.fit_transform(X)
```

Стандартизация приводит значения признаков к виду стандартного нормального распределения: значения будут центрированы относительно 0, а средний разброс данных составит 1. 

###   Создание матрицы расстояний функцией linkage() на стандартизованной матрице признаков и построение дендрограммы. Количество кластеров, которые можно выделить.

```python
linked = linkage(x_sc, method = 'ward') 
```

В переменной linked сохранена таблица «связок» между объектами. Используем “ward” в качестве метода, т.к. он позволяет он минимизировать варианты расстояний между кластерами.

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация')
plt.show() 
```

Предложенное оптимальное число кластеров равняется 4 (разные цвета на графике). При этом фиолетовый кластер существенно больше остальных. Его можно разделить на 2 кластера. Далее в работе будут использоваться 5 кластеров.

###    Обучение модели кластеризации на основании алгоритма K-Means и прогнозирование кластеров клиентов.

```python
# задаём модель k_means с числом кластеров 5 и фиксируем значение random_state
km = KMeans(n_clusters=5, random_state=0)
```

```python
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(x_sc)
```

В переменной labels сохраняются индексы предложенных алгоритмом групп. 

```python
# сохраняем метки кластера в поле датасета
df['churn_km'] = labels
```

###    Изучение средних значений признаков для кластеров.

```python
# выводим статистику по средним значениям признаков по кластеру
df.groupby('churn_km').mean()
```

Характерной особенностью кластеров является следующее. 

У нулевой группы самый низкий отток, договора заключаются чаще на 1 год. И в этой группе самое большое количество клиентов, пришедших по рекомендации знакомых, а также воспользовавшиеся скидками от компании-работодателя.

Клиенты первой группы не оставили контактный номер телефона. Но данный признак не оказывают особого влияния на остальные факторы.

Во второй группе клиенты дальше всего расположены от фитнес-клуба. Они реже приходят по рекомендациям знакомых и реже посещают групповые занятия, т.к., по видимому, меньше стимула далеко идти или ехать. Отток, при этом, достаточно высокий - 0,44.

В третьей группе расположились клиенты с самым высоким оттоком - 0,51. Здесь сконцентрировались клиенты с самыми краткосрочными абонементами (среднее - 1,9 мес.) несмотря на то, что живут или работают ближе всех к фитнес-центру. 

В четвертой группе расположились клиенты, которые чаще посещают фитнес-центр в течение недели - 2,85 раза. Хотя абонементы они предпочитают краткосрочные, тем не менее, отток у них сравнительно низкий - 0,07. Личная мотивация заниматься в фитнес-клубе является для них самым сильным фактором, невзирая на срок абонемента.
