#  Изучение закономерностей, определяющих успешность игр

###### Описание проекта
Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

###### Описание данных
* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

## Проверка гипотез

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшенигры) и Sports (англ. «спортивные соревнования») разные.

<p>Введем основную и альтернативные гипотезы:</p>

$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средние пользовательские рейтинги платформ Xbox One и PC одинаковые}\\
   H_1 :\text{Средние пользовательские рейтинги платформ Xbox One и PC различаются}
 \end{cases}
\end{equation*}$

```python
# средний рейтинг
rating_XOne = df_since_2014_with_scores.loc[df_since_2014_with_scores['platform'] == "XOne"]['user_score']
rating_XOne.mean()
```

```python
# средний рейтинг
rating_PC = df_since_2014_with_scores.loc[df_since_2014_with_scores['platform'] == "PC"]['user_score']
rating_PC.mean()
```

```python
# дисперсия Xone
variance_estimate_XOne = np.var(df_since_2014_with_scores.loc[df_since_2014_with_scores['platform'] == "XOne"]['user_score'], ddof=1)
variance_estimate_XOne
```

```python
# стандартное отклонение Xone
standard_deviation_Xone = np.sqrt(variance_estimate_XOne)
standard_deviation_Xone
```

```python
# дисперсия PC
variance_estimate_PC = np.var(df_since_2014_with_scores.loc[df_since_2014_with_scores['platform'] == "PC"]['user_score'], ddof=1)
variance_estimate_PC
```

```python
# стандартное отклонение PC
standard_deviation_PC = np.sqrt(variance_estimate_PC)
standard_deviation_PC
```

```python
# функция гистограмм частоты распределения
for param in ['user_score']:
    sns.distplot(df_since_2014_with_scores.loc[df_since_2014_with_scores['platform'] == "XOne", param], label='XOne / '+param)
    sns.distplot(df_since_2014_with_scores.loc[df_since_2014_with_scores['platform'] == "PC", param], label='PC / '+param)
    
    plt.ylabel('частота')
    plt.xlabel(param)
    plt.grid()
    plt.title('Пользовательские рейтинги')
    plt.legend()
    plt.show()
```

```python
# qq-plot для оценки соответствия распределения наборов значений стандартному нормальному по выручке
plt.subplot(2,2,1)
st.probplot(rating_XOne, dist="norm", plot=plt)
plt.subplot(2,2,2)
st.probplot(rating_PC, dist="norm", plot=plt)
plt.show()
```

```python
# расчет p-value для сравнения средних двух выборок
results = st.mannwhitneyu(rating_XOne, 
                          rating_PC, alternative='two-sided')
print('p-value: ', results.pvalue)

alpha = 0.05

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# расчет p-value для сравнения средних двух выборок
results = st.ttest_ind(rating_XOne, 
                          rating_PC)
print('p-value: ', results.pvalue)

alpha = 0.05

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

В данном исследовании пороговое значение уровня значимости (p-value) равняется 0,05, что является общепринятым значением. Нулевая гипотеза отвергается, если p-value меньше 0,05.

Стандартное отклонение по Xbox One составило 1,32, по PC - 1,66. Распределение случайных величин на графике имеет небольшую скошенность влево. Графики квантиль-квантиль также немного отклоняются от нормального распределения. В связи с этим в исследовании использовался непараметрический тест Манна-Уитни. Для сравнения также проверили гипотезу с помощью t-теста, который используется при нормальном распределении. Оба теста показали результат уровня значимости в 0,33 и 0,13 соответственно. В результате, нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC не удалось отвергнуть.
```python
# расчет p-value для сравнения средних двух выборок
results = st.ttest_ind(rating_Action, 
                          rating_Sports)
print('p-value: ', results.pvalue)

alpha = 0.05

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
