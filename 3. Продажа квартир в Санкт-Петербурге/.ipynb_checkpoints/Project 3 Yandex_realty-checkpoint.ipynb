## Предобработка данных

### Обработка пропусков

```python
# подсчёт пропусков
data.isna().sum()
```

```python
data.isna().mean().sort_values(ascending=False).to_frame().rename(columns={0 : 'доля_пропусков'}).style.format({'доля_пропусков' : '{:.2%}'.format})
```

```python
# наличие дубликатов в таблице
data.duplicated().sum()
```

```python
# замена пропущенных значений столбца на 0
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].isna().sum()
```

```python
# замена пропущенных значений столбца на False
data['is_apartment'] = data['is_apartment'].fillna(False)
data.is_apartment.value_counts()
```

```python
# (старый вариант)
# замена пропущенных значений столбца жилплощади на медиану
# data['living_area'] = data['living_area'] \
#    .fillna(data.groupby(['total_area'])['living_area'].transform('median'))
# data['living_area'].isna().sum()
```

```python
# замена пропущенных значений столбца жилплощади на медиану
data['living_area'] = data['living_area'] \
    .fillna(data.groupby(['rooms'])['living_area'].transform('median'))
data['living_area'].isna().sum()
```

```python
# замена пропущенных значений столбца высоты потолков на медиану
data['ceiling_height'] = data['ceiling_height'] \
    .fillna(data.groupby(['total_area'])['ceiling_height'].transform('median'))
data['ceiling_height'].isna().sum()
```

```python
# замена пропущенных значений столбца площади кухни на медиану
kitchen_area_median = data['total_area'].median() - data['living_area'].median()
data['kitchen_area'].fillna(kitchen_area_median, inplace=True)
data['kitchen_area'].isna().sum()
```

```python
# перевод значений столбца из строки в дату
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],
format='%Y-%m-%d')
```

```python
# гистограмма для столбца first_day_exposition для среза по isna() по days_exposition
import warnings
warnings.filterwarnings("ignore")

data[data.days_exposition.isna()]['first_day_exposition'].hist(figsize=(12,6))
plt.xticks(rotation=15)
plt.show()
```

```python
# гистограмма для столбца first_day_exposition для среза по notna() по days_exposition
data[data.days_exposition.notna()]['first_day_exposition'].hist(figsize=(12,6))
plt.xticks(rotation=15)
plt.show()
```

## Расчёты и добавление результатов в таблицу

```python
# добавление столбца: цена квадратного метра
data['price_of_m2'] = data.last_price / data.total_area
```

```python
# добавление столбцов: месяц и год публикации объявления
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
```

```python
# добавление столбца: день недели публикации объявления
data['day_of_week'] = data['first_day_exposition'].apply(lambda x: x.isoweekday())
```

```python
# преобразование значений столбца в дни недели на русском языке
list_of_days = {1: 'понедельник', 2: 'вторник', 3: 'среда', 4: 'четверг', 
                    5: 'пятница', 6: 'суббота', 7: 'воскресенье'}
data['day_of_week'] = data['day_of_week'].map(list_of_days)
```

```python
# проверка новых столбцов
data[['day_of_week', 'year', 'month']].head(10)
```

```python
# добавление столбца: этаж квартиры; варианты — первый, последний, другой
def floor_category(floor):
    for row in data.floor:
        if floor == 1: 
            return "первый"
        elif floor == row: 
            return "последний"
        return "другой"

# создание нового столбца
data['floor_category'] = data['floor'].apply(floor_category)
data.floor_category.unique()
```

```python
# добавление столбца: соотношение жилой и общей площади, а также отношение площади кухни к общей
data['living_total_areas_ratio'] = data.living_area / data.total_area
data['kitchen_total_areas_ratio'] = data.kitchen_area / data.total_area
data.head()
```

*    этаж квартиры; варианты — первый, последний, другой;

### Факторы, которые больше всего влияют на стоимость квартиры. Изучение зависимости цены от площади, числа комнат, удалённости от центра. Изучение зависимости цены от того, на каком этаже расположена квартира: первом, последнем или другом. Также изучение зависимости от даты размещения: дня недели, месяца и года.

```python
data_new.total_area.describe()
```

```python
# функция для определения категории общей площади
def total_area_cat(area):
    if area <= 39: 
        return "небольшая"
    elif area <= 48.4: 
        return "ниже среднего"
    elif area <= 62.2: 
        return "выше среднего"
    else: 
        return "большая"

# создание нового столбца с категориями общей площади
data_new['total_area_cat'] = data_new['total_area'].apply(total_area_cat)
data_new['total_area_cat'].value_counts()

# игнорирование предупреждений
import warnings
warnings.filterwarnings("ignore")
```

```python
# создание таблицы об общей площади и цене за кв. м по категориям
price_m2_and_total_area_cat = data_new.groupby('total_area_cat')['price_of_m2'] \
                                  .agg(['count', 'sum', lambda x: '{:.8}' \
                                  .format(x.mean())])
price_m2_and_total_area_cat.columns = ['count', 'sum', 'mean']
price_m2_and_total_area_cat
```

С 2015 г. цены на жилье повышаются каждый год, как по всей выборке, так и в центре города. Расстояние до центра не играет роли, т.к. уже была осуществлена выборка квартир в зависимости от стоимости жилья, поэтому высокая цена примерно одинакова в этом районе. Идеальнее всего приобретать жилье в июне, когда продавцы устанавливают самую низкую цену как в центре, так и в других районах. В центре Санкт-Петербурга самые высокие цены в марте (120,9 тыс. руб.). В остальных районах - в августе и сентябре (92,9 и 92,7 тыс. руб.). Сравнивая цену за 1 кв. метр жилья в зависимости от дня публикации можно обнаружить, что, в целом, продавцы устанавливают чуть более высокие цены в будни, а в центре Санкт-Петербурга - в выходные.
